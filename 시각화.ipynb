### **축 지정하기**

* ticks : 축에 적히는 수 위치

* ticklabels : 축의 이름

```python
fig=plt.figure(figsize=(12,8))
ax1=fig.add_subplot(1,3,1)
ax1.plot([1,1,1],label='1')  
ax1.plot([2,2,2],label='2')
ax1.plot([3,3,3],label='3')

ax1.set_xticks([0,1,2])        # x 좌표 지정

ax1.text(x=1,y=2,s='This is Text')

ax1.legend()

ax2=fig.add_subplot(1,3,2)
ax2.plot([1,1,1],label='one')  
ax2.plot([2,2,2],label='two')
ax2.plot([3,3,3],label='three')

ax2.set_xticklabels(['one','two','three'])        # x 좌표 지정 -> 원하는 위치에서 볼 수 없음

ax2.annotate(s='This is text',xy=(1,2))
ax2.legend()

ax3=fig.add_subplot(1,3,3)
ax3.plot([1,1,1],label='one')  
ax3.plot([2,2,2],label='two')
ax3.plot([3,3,3],label='three')

ax3.set_xticks([0,1,2])
ax3.set_xticklabels(['one','two','three'])        # xticks하고 같이 써야 원하는 위치에 라벨을 붙일 수 있음

ax3.annotate(s='This is text',xy=(1,2),xytext=(1.2,2.2),arrowprops=dict(facecolor='black'))     # 화살표 추가
ax3.legend()

plt.show()
```

### Stacked Bar Plot

<p>
<div>

* 2개 이상의 그룹을 쌓아서 표현

* 각 bar에서 나타나는 그룹의 순서는 유지

* 맨 밑은 파악 쉽지만 그 외 분포는 파악하기 어려움

* bar()에서는 **bottom** 파라미터 사용

* barh()에서는 **left** 파라미터 사용




```python
fig, axes=plt.subplots(1,2,figsize=(15,7))

group_cnt=student['race/ethnicity'].value_counts().sort_index()
axes[0].bar(group_cnt.index,group_cnt,color='darkgray')
axes[1].bar(group['male'].index,group['male'],color='royalblue')
axes[1].bar(group['female'].index,group['female'],bottom=group['male'],color='tomato')
plt.show()
```

```python
# Percentage Stacked Bar Plot

fig, ax=plt.subplots(1,1,figsize=(15,7))

group=group.sort_index(ascending=False)
total=group['male']+group['female']
ax.barh(group['male'].index, group['male']/total,color='royalblue')
ax.barh(group['female'].index, group['female']/total, left=group['male']/total, color='tomato')

plt.show()
```

```python
# 여백 없애기

fig, ax=plt.subplots(1,1,figsize=(15,7))

group=group.sort_index(ascending=False)
total=group['male']+group['female']
ax.barh(group['male'].index, group['male']/total,color='royalblue')
ax.barh(group['female'].index, group['female']/total, left=group['male']/total, color='tomato')

ax.set_xlim(0,1)
for s in ['top','bottom','left','right']:
  ax.spines[s].set_visible(False)
plt.show()
```

### Overlapped Bar Plot
* 겹처서 그리는 방식
* 같은 축에서 사용해서 비교가 쉬움
* 투명도로 겹치는 부분 파악(**alpha**)
* Bar plot보다 Area plot에서 더 효과적

![image](https://user-images.githubusercontent.com/63588046/152275169-494f4d9c-b4a8-4f01-8eb4-b4434de501db.png)

```python
group=group.sort_index()

fig, axes=plt.subplots(2,2,figsize=(15,15))
axes=axes.flatten()       # 이거 안 넣으면 numpy 에러

for idx,alpha in enumerate([1,0.7,0.5,0.3]):
  axes[idx].bar(group['male'].index,group['male'],color='royalblue',alpha=alpha)
  axes[idx].bar(group['female'].index,group['female'],color='tomato',alpha=alpha)
  axes[idx].set_title=(f'Alpha = {alpha}')

for ax in axes:
  ax.set_ylim(0,200)

plt.show()
```

### Grouped Bar Plot
* 그룹별 범주에 따른 bar를 이웃하게 배치
* set_xticks(), set_xticklables() 사용
* x축이 0,1,2,3로 시작한다면

=> 한 그래프는 0-width/2, 1-width/2 ...

=> 한 그래프는 0+width/2, 1+width/2 ...

![image](https://user-images.githubusercontent.com/63588046/152275199-1c05b304-3fba-43b7-a8af-10df658857f9.png)

* 2개 : -1/2, 1/2
* 3개 : -1, 0, 1
* 4개 : -3/2, -1/2, 1/2, 3/2

=> (N-1)/2 개씩 차이가 남

```python
fig, ax=plt.subplots(1,2, figsize=(12,7))
idx=np.arange(len(group['male'].index))
width=0.35

ax[0].bar(idx-width/2, group['male'],color='royalblue',width=width)
ax[0].bar(idx+width/2, group['female'],color='tomato',width=width)

ax[1].bar(idx-width/2, group['male'],color='royalblue',width=width)
ax[1].bar(idx+width/2, group['female'],color='tomato',width=width)
ax[1].set_xticks(idx)
ax[1].set_xticklabels(group['male'].index)

plt.show()
```

## 정확한 LIne Plot
### 추세에 집중
* Bar Plot과 다르게 반드시 축을 0에 초점 둘 필요 X
* 구체적X (디테일은 표로)
* 생략되지 않는 선에서 범위 조정하여 변화율 관찰(**.set_ylim()**)

### 간격
* 규칙적 간격이 아닌 경우 점으로 표시하여 오해를 줄이자
* 보간(점과 점 사이 이음)은 좋은 방법일 수 있지만 없는 데이터를 있다고 생각 할 수 있음

  ->일반적인 분석에서는 지양

### 이중 축
* 한 plot에 대해 2개의 축 사용
* 같은 시간 축에 대해 서로 다른 종류의 데이터 표현할 때 **.twinx()** 사용
* 한 데이터의 다른 단위 사용(**.secondary_xaxis(), secondary_yaixis()**)



```python
# 이중축
apple=stock[stock['symbol']=='AAPL']
google=stock[stock['symbol']=='GOOGL']

color='royalblue'

fig,ax=plt.subplots(1,1,figsize=(15,8),dpi=300)       #dpi : 해상도
ax.plot(google.index,google['close'],color=color)
ax.set_xlabel('date')
ax.set_ylabel('close price',color=color)
ax.tick_params(axis='y',labelcolor=color)

ax2=ax.twinx()
color='tomato'
ax2.plot(google.index,google['volume'],color=color)
ax2.set_ylabel('volume',color=color)
ax2.tick_params(axis='y',labelcolor=color)

ax.set_title('google close price & volume', loc='left', fontsize=15)
plt.show()
```

```python
iris=pd.read_csv('/content/drive/MyDrive/부스트캠프/시각화/Iris.csv')
iris.head()
```

```python
iris.describe(include='all')
```

```python
fig=plt.figure(figsize=(7,7))
ax=fig.add_subplot(1,1,1)
ax.scatter(x=iris['SepalLengthCm'],y=iris['SepalWidthCm'])
plt.show()
```

```python
fig=plt.figure(figsize=(7,7))
ax=fig.add_subplot(1,1,1)
alc_mean=iris['SepalLengthCm'].mean()
awc_mean=iris['SepalWidthCm'].mean()
ax.scatter(x=iris['SepalLengthCm'],y=iris['SepalWidthCm'],
           c=['royalblue' if xx<=alc_mean else 'gray' for xx in iris['SepalLengthCm']])
plt.show()
```

```python
# 꽃의 종류에 따라 다르게 분류
fig=plt.figure(figsize=(14,7))
ax1=fig.add_subplot(1,2,1)
for species in iris['Species'].unique():
  iris_sub=iris[iris['Species']==species]
  ax1.scatter(x=iris_sub['SepalLengthCm'],
             y=iris_sub['SepalWidthCm'],
             label=species)
ax1.legend()

ax2=fig.add_subplot(1,2,2)
for species in iris['Species'].unique():
  iris_sub=iris[iris['Species']==species]
  ax2.scatter(x=iris_sub['PetalLengthCm'],
             y=iris_sub['PetalWidthCm'],
             label=species)
  
ax2.axvline(2.6,color='gray',linestyle=':')   # 흑색 실선 그리기
ax2.axhline(0.8,color='gray',linestyle=':')  

ax2.legend()

plt.show()
```

```python
# 엄청 여러개 그리기 - 이후 seaborn에서 자세히 설명
fig,axes=plt.subplots(4,4,figsize=(14,14))
feat=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']

for i,f1 in enumerate(feat):
  for j, f2 in enumerate(feat):
    if i<=j:
      axes[i][j].set_visible(False)
      continue
    for species in iris['Species'].unique():
      iris_sub=iris[iris['Species']==species]
      axes[i][j].scatter(x=iris_sub[f2],
                         y=iris_sub[f1],
                         label=species,
                         alpha=0.7)
    if i==3: axes[i][j].set_xlabel(f2)
    if j==0: axes[i][j].set_ylabel(f1)
plt.tight_layout()
plt.show()
```
